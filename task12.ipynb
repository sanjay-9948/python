```python
import pygame, sys

pygame.init()

w = pygame.display.set_mode((500, 300))
p = pygame.Rect(50, 130, 40, 40)
e = pygame.Rect(400, 130, 40, 40)

c = pygame.time.Clock()
s = 5   # speed
running = True # Flag to control the game loop

while running:
    for ev in pygame.event.get():
        if ev.type == pygame.QUIT:
            running = False  # Set flag to False to exit the loop

    k = pygame.key.get_pressed()
    p.x += (k[pygame.K_RIGHT] - k[pygame.K_LEFT]) * s
    p.y += (k[pygame.K_DOWN] - k[pygame.K_UP]) * s

    # Enemy auto movement (optional â€“ based on your notes)
    e.x -= 4
    if e.x <= 0:
        e.x = 500

    w.fill((255, 255, 255))

    pygame.draw.rect(w, (0, 0, 255), p)
    pygame.draw.rect(w, (255, 0, 0), e)

    if p.colliderect(e):  # collision check
        font = pygame.font.SysFont(None, 50)
        text = font.render("Game Over!", True, (0, 0, 0))
        w.blit(text, (150, 120))
        pygame.display.update()
        pygame.time.delay(2000)
        running = False # Set flag to False to exit the loop after collision

    pygame.display.update()
    c.tick(30)

pygame.quit() # Quit Pygame after the loop finishes
```

```python
# Text-based game simulation in Colab

player_pos = 50
enemy_pos = 400
speed = 5

def show_status():
    print(f"Player position: {player_pos}")
    print(f"Enemy position: {enemy_pos}")
    print("-" * 30)

def move_player(direction):
    global player_pos
    if direction == "right":
        player_pos += speed
    elif direction == "left":
        player_pos -= speed

def move_enemy():
    global enemy_pos
    enemy_pos -= 4
    if enemy_pos <= 0:
        enemy_pos = 500

def check_collision():
    return abs(player_pos - enemy_pos) < 40

# Game loop
for turn in range(20):
    show_status()
    move = input("Move (left/right/none): ").strip().lower()
    move_player(move)
    move_enemy()

    if check_collision():
        print("ðŸ’¥ Game Over! You collided with the enemy.")
        break
else:
    print("âœ… You survived 20 turns!")
```
